```python
data1.rename(columns={'from_name':'key'}, inplace=True)
data2.rename(columns={'name':'key'}, inplace=True)
data2
```

```python
data = pd.merge(data1, data2, on='key')
data
```

```python

json_data = []
import json
for row in data.iterrows():
    tmp = {
        'source': 'None',
        'target': 'None',
        'relation': 'None',
        'value': 'None',
    }
    tmp['source'] = row[1]['key']
    tmp['target'] = row[1]['to_name']
    tmp['relation'] = row[1]['property']
    tmp['value'] = row[1]['重要程度']
    print(tmp)
    json_data.append(tmp)
```

```python
json_data
```

```python
nodes = []
s = set()
for row in data.iterrows():
    s.add(row[1]['key'])
    s.add(row[1]['to_name'])
for i in s:
    tmp = {
        'class': '精品特色菜',
        'group': '1',
        'size' : '10',
        'id' : i
    }
    print(tmp)
    nodes.append(tmp)
print(len(s)) 
```

```python
ress = {
    'links':json_data,
    'nodes':nodes
}
ress
```

```python
with open('java.json', 'w', encoding='utf-8') as file: 
    json.dump(ress,file ,ensure_ascii=False)
```
